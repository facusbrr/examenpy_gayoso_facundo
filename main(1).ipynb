### Actividad 3

- 3.1: El tratamiento de valores nulos: rellenamos con la media
- 3.2: Convertimos la columna fecha a datetime --> Algunas fechas estan distintas, por eso NaT.
  - A los valores NaT se los elimino directamente.
- 3.3: Se agregaron mayusculas en nombre y apellido.
  - A los valores NaN se le agrego un valor por defecto.
- 3.4: En el caso de votos se convirtieron los valores a mayusculas
- 3.5: El DNI se paso a tipo INT

```python
columnas = votos_df.select_dtypes(include=[np.number]).columns
votos_df[columnas] = votos_df[columnas].fillna(votos_df[columnas].mean())
```

```python
if 'fecha_votacion' in votos_df.columns:
    votos_df['fecha_votacion'] = pd.to_datetime(votos_df['fecha_votacion'], errors='coerce')
    votos_df = votos_df.dropna(subset=['fecha_votacion'])
```

```python
if 'nombre' in votos_df.columns:
    votos_df['nombre'] = votos_df['nombre'].fillna('No encontrado')
if 'apellido' in votos_df.columns:
    votos_df['apellido'] = votos_df['apellido'].fillna('No encontrado')
if 'voto' in votos_df.columns:
    votos_df['voto'] = votos_df['voto'].fillna('NULO')
if 'provincia' in votos_df.columns:
    votos_df['provincia'] = votos_df['provincia'].fillna('No encontrado')
```

```python
if 'nombre' in votos_df.columns:
    votos_df['nombre'] = votos_df['nombre'].str.title()
if 'apellido' in votos_df.columns:
    votos_df['apellido'] = votos_df['apellido'].str.title()
```

```python
if 'voto' in votos_df.columns:
    votos_df['voto'] = votos_df['voto'].str.upper()
```

```python
if 'dni' in votos_df:
    votos_df['dni'] = pd.to_numeric(votos_df['dni'], errors='coerce')
    votos_df['dni'] = votos_df['dni'].round().astype('Int64')
```
